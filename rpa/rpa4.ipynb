```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

# 크롬 드라이버 : 크롬 웹 브라우저를 동작하기 위해 필요한 프로그램
# selenium 4.6 버전부터 크롬 드라이버 별도 다운로드가 필요 없음

driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://shopping.naver.com/ns/home")
time.sleep(0.5)

# 상품명 입력 엔터
search = driver.find_element(By.CLASS_NAME, "_searchInput_search_text_83jy9")
input_text = "후드집업"
for char in input_text:
    search.send_keys(char)
    time.sleep(0.1)
search.send_keys(Keys.ENTER)
time.sleep(2)

# 페이지 이동 후 스크롤 이동 : window.scrollTo(시작위치, 종료위치)
# driver.execute_script("window.scrollTo(0, 2160)")

# 1. 현재 문서 높이
prev_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # 현재 가져온 페이지의 높이만큼 스크롤 이동
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    time.sleep(1)

    # 스크롤이 진행된 후 문서 위치
    curr_height = driver.execute_script("return document.body.scrollHeight")

    # 3. 현재 스크롤된 위치와 초기화된 문서의 위치가 같다면 중지
    if curr_height == prev_height:
        break

    # 2. 현재 문서 높이는 스크롤이 진행된 후 높이로 초기화
    prev_height = curr_height

# 맨 처음으로 가기
driver.execute_script("window.scrollTo(0, 0)")

time.sleep(3)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pyautogui as pag
from openpyxl import Workbook
from openpyxl.drawing.image import Image
import requests
from fake_useragent import UserAgent
from io import BytesIO

wb = Workbook()
ws = wb.active

ws.append(["번호", "제품명", "가격", "이미지"])
ws.column_dimensions["A"].width = 12
ws.column_dimensions["B"].width = 50
ws.column_dimensions["C"].width = 14
ws.column_dimensions["D"].width = 20

driver = webdriver.Chrome()
driver.maximize_window()

# 다나와 사이트 접속
driver.get("https://prod.danawa.com/list/?cate=112758&15main_11_02=")
time.sleep(1.5)

# 제조사별 클릭하기
driver.find_element(By.XPATH, "//*[@id='dlMaker_simple']/dd/div[2]/button[1]").click()
time.sleep(1)

# Apple 클릭하기
driver.find_element(By.CSS_SELECTOR, "#selectMaker_simple_priceCompare_A > li:nth-child(15) > label").click()
time.sleep(3)

cur_page, target_crawl_num = 1,5
idx = 1
search = driver.find_element(By.ID, "AKCSearch")
while cur_page <= target_crawl_num:

    print(f"현재 페이지 : {cur_page}")
    search.clear()
    input_text = f"{cur_page}페이지 정보 찾아오는 중"
    for char in input_text:
        search.send_keys(char)
        time.sleep(0.02)
    

    # 데이터 수집
    # 제품명, 가격, 이미지 소스
    items = driver.find_elements(By.CLASS_NAME, "prod_item")
    for item in items:
        title = item.find_element(By.CSS_SELECTOR,".prod_name > a").text
        price = item.find_element(By.CSS_SELECTOR, "p.price_sect > a").text
        src = item.find_element(By.CSS_SELECTOR, ".thumb_link img")

        if src.get_attribute("data-original"):
            image_src = src.get_attribute("data-original")
        else:
            image_src = src.get_attribute("src")

        if "https:" not in image_src:
            image_src = "https:"+image_src

        print(idx, title, price, image_src)
        # 번호 제품명 가격 이미지
        ws.append([idx, title, price])
        
        idx += 1

        # 이미지를 가져와서 엑셀에 저장
        res = requests.get(image_src, headers={"user-agent":UserAgent().chrome})
        save_img = BytesIO(res.content)
        
        ws.row_dimensions[idx].height = 87
        img = Image(save_img)
        img.width = 150
        img.height = 120
        ws.add_image(img, "D"+str(idx))

    cur_page += 1

    if cur_page > target_crawl_num:
        search = driver.find_element(By.ID, "AKCSearch")
        search.clear()
        input_text = "가져오기 완료"
        for char in input_text:
            search.send_keys(char)
            time.sleep(0.05)
        break


    
    # 다음 페이지 번호 클릭하기
    driver.find_element(By.CSS_SELECTOR, f"div.number_wrap > a:nth-child({cur_page})").click()
    # productListArea > div.prod_num_nav > div > div > a:nth-child(2)
    # productListArea > div.prod_num_nav > div > div > a:nth-child(3)
    time.sleep(2)

time.sleep(0.05)

search.clear()
input_text = "액셀 저장 중"
for char in input_text:
    search.send_keys(char)
    time.sleep(0.05)
    
time.sleep(1)
    
wb.save("./다나와_애플_노트북.xlsx")
wb.close()

time.sleep(1)

search.clear()
input_text = "액셀 저장 완료"
for char in input_text:
    search.send_keys(char)
    time.sleep(0.05)

time.sleep(0.5)

driver.quit()
```

```python
# DB 저장

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pyautogui as pag
import re
import sqlite3

conn = sqlite3.connect("./danawa.db", isolation_level=None)
cursor = conn.cursor()

driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://prod.danawa.com/list/?cate=112758&15main_11_02=")
time.sleep(1.5)

# 제조사별 클릭하기
driver.find_element(By.XPATH, "//*[@id='dlMaker_simple']/dd/div[2]/button[1]").click()
time.sleep(1)

# Apple 클릭하기
driver.find_element(By.CSS_SELECTOR, "#selectMaker_simple_priceCompare_A > li:nth-child(15) > label").click()
time.sleep(3)

cur_page, target_crawl_num = 1,5
idx = 1
search = driver.find_element(By.ID, "AKCSearch")
while cur_page <= target_crawl_num:

    print(f"현재 페이지 : {cur_page}")
    search.clear()
    input_text = f"{cur_page}페이지 정보 찾아오는 중"
    for char in input_text:
        search.send_keys(char)
        time.sleep(0.02)
    

    # 데이터 수집
    # 제품명, 가격, 이미지 소스
    items = driver.find_elements(By.CLASS_NAME, "prod_item")
    for item in items:
        title = item.find_element(By.CSS_SELECTOR,".prod_name > a").text
        price = item.find_element(By.CSS_SELECTOR, "p.price_sect > a > strong").text
        src = item.find_element(By.CSS_SELECTOR, ".thumb_link img")

        if src.get_attribute("data-original"):
            image_src = src.get_attribute("data-original")
        else:
            image_src = src.get_attribute("src")

        if "https:" not in image_src:
            image_src = "https:"+image_src

        print(idx, title, price, image_src)
        # DB 저장
        price = re.sub(",","",price) # price -> integer로 저장하기 위해 , 떼어내기
        sql = "INSERT INTO appleStuff(title, price, image_source) VALUES(?,?,?)"
        cursor.execute(sql,(title,price,image_src))
        
        idx += 1

    cur_page += 1

    if cur_page > target_crawl_num:
        search = driver.find_element(By.ID, "AKCSearch")
        search.clear()
        input_text = "가져오기 완료"
        for char in input_text:
            search.send_keys(char)
            time.sleep(0.05)
        break


    # 다음 페이지 번호 클릭하기
    driver.find_element(By.CSS_SELECTOR, f"div.number_wrap > a:nth-child({cur_page})").click()
    # productListArea > div.prod_num_nav > div > div > a:nth-child(2)
    # productListArea > div.prod_num_nav > div > div > a:nth-child(3)
    time.sleep(2)


time.sleep(1)

driver.quit()
```
