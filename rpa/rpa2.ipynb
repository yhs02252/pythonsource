```python
import pyautogui as pag

# 해상도 확인
size = pag.size()
print(size)
print("width =", size[0])
```

```python
# 마우스 조작

# 마우스 좌표 가져오기
position = pag.position()
print(position)
print(position.x, position.y)
```

```python
# 마우스 이동(절대좌표) : moveTo(x, y, 움직이는 시간) - (스크린 기준)
pag.moveTo(100,100,0.5)
pag.moveTo(200,200,0.5)
pag.moveTo(300,300,0.5)
```

```python
# 마우스 이동(상대좌표) :  move(x, y, 움직이는 시간) or moveRel() - (마우스 포인터 기준)
pag.moveTo(100,100,0.5)
pag.move(300,300,0.25)
print(pag.position())
pag.move(100,100,0.5)
```

```python
# 마우스 클릭(왼쪽) : click()

# 현재위치에서 클릭
pag.click()

# 특정 위치에서 클릭
# pag.click(x축, y축)
```

#### screenshot

```python
# 스크린 샷
# 전체 화면 : pag.screenshot("파일명")
# pag.screenshot("./screen.png")

# 특정 위치에 대한 스크린 샷 : screenshot("파일명", region=(left, top, width, height))
# region=(left, top, width, height) - 왼쪽 위 끝자락 기준 오른쪽으로 width, 아래로 height
pag.sleep(3)
pag.screenshot("./screenshot.png", region=(1409,489,316,530))
```

```python
# 스크린샷과 같은 이미지 위치 찾기
# confidence = 수치 : 이미지 일치율
print(pag.locateOnScreen("./screenshot.png"))
```

```python
# vscode File 메뉴 캡쳐
pag.screenshot("./file.png", region=(0,35,45,30))

pag.sleep(2)

print(pag.locateOnScreen("./file.png"))
```

```python
file_locate = pag.locateOnScreen("./file.png",confidence=0.9)
pag.click(file_locate)
```

```python
# locateAllOnScreen()
pag.sleep(1)
pag.screenshot("./checkbox.png", region=(1295, 667, 76, 22))
```

```python
for i in pag.locateAllOnScreen("./checkbox.png", confidence=0.9):
    print(i)
    pag.click(i, interval=0.5)
```

```python
# 윈도우
# 현재 활성화된 창
window = pag.getActiveWindow()
print(window)
print(window.title)
print(window.top, window.left, window.width, window.height)
```

```python
# 활성화된 창 전부 가져오기
for w in pag.getAllWindows():
    print(w)
```

```python
# 특정 창 가져오기
for w in pag.getWindowsWithTitle("새 텍스트 문서"):
    print(w)
```

```python
# 리스트 형태로 찾은 결과물 index 붙혀서 실행
w = pag.getWindowsWithTitle("새 텍스트 문서")[0]
if not w.isActive:
    w.activate
```

```python
# 창 최대화
if not w.isMaximized:
    w.maximize()
```

```python
# 창 최소화
if not w.isMinimized:
    w.minimize()
```

## 키보드 조작

```python
# 키보드 조작

w = pag.getWindowsWithTitle("새 텍스트 문서")[0]
w.activate()

pag.write("write")
```

```python
# 키보드 조작

w = pag.getWindowsWithTitle("새 텍스트 문서")[0]
w.activate()

pag.write("pyautogui", interval=0.25)
```

```python
# 키보드 조작

w = pag.getWindowsWithTitle("새 텍스트 문서")[0]
w.activate()

pag.write("안녕하세요", interval=0.25) # 한글 안됨
```

```python
import pyperclip

w = pag.getWindowsWithTitle("새 텍스트 문서")[0]
w.activate()

# 복사 + 붙혀넣기 처리
pyperclip.copy("안녕하세요") # 클립보드
pag.hotkey("ctrl", "v") # ctrl + v
```

```python
def hangul_write(text):
    pyperclip.copy(text)
    pag.hotkey("ctrl","v")
```

```python
w = pag.getWindowsWithTitle("새 텍스트 문서")[0]
w.activate()
hangul_write("반갑습니다")
hangul_write("hello")
```
