```python
# score.txt 에 있는 점수를 읽어온 후 합계와 평균을 구한 후 출력
f = open("./score.txt", "r", encoding="utf-8")
scores = []
for i in f:
    scores.append(int(i))

print(f"평균 : {sum(scores) / len(scores)}")
```

```python
# sample.txt 읽은 후 합계와 평균 구하기
# result.txt 작성(합계, 평균 기록)

f = open("./sample.txt", "r", encoding="utf-8")
sampleScores= []
for n in f:
    sampleScores.append(int(n))

r = open("./result.txt", "w", encoding="utf-8")
r.write(f"sampleScore 숫자의 합계는  : {sum(sampleScores)}")
r.write("\n")
r.write(f"sampleScore 숫자의 평균은  : {sum(sampleScores)/len(sampleScores)}")
r.close()
```

```python
# info.txt 읽기
info = open("./info.txt", "r", encoding="utf-8")
infoList = []
for i in info:
    name,weight,height = i.strip().split(", ")

    bmi = int(weight) / (int(height) / 100) ** 2


    if 25 <= bmi:
        result = "과체중"
    elif 18.5 <= bmi:
        result = "정상체중"
    else :
        result = "저체중"

    print("정보 출력 : " + name,weight,height)
    print(f"이름 : {name}\n")
    print(f"몸무게 : {weight}\n")
    print(f"키 : {height}\n")
    print(f"BMI : {bmi}\n")
    print(f"결과 : {result}\n")

# BMI 지수 계산 후 화면에 출력
# [출력결과]
# 타차, 66, 158
# 이름 : 타차
# 몸무게 : 66 
# 키 : 158
# BMI : 33.156 (BMI = weight / (height / 100) ** 2)
# 결과 : 과체중
```

```python
while True :
    no = int(input("1. 암호화 | 2. 복호화 | 3. 종료" ))
    if no == 1:
        # origin 읽고 암호화 후 encryption.txt 로 저장
        origin = open("./origin.txt", "r", encoding="utf-8")
        content = origin.read()
        
        encryption = open("./encryption.txt", "w",encoding="utf-8")
        for c in content:
            encryption.write(chr(ord(c) + 100))
        
    elif no == 2:
        # encryption.txt 를 읽은 후 원래 문자대로 화면 출력
        encryption2 = open("./encryption.txt", "r",encoding="utf-8")
        content = encryption2.read()
        for i in range(len(content)):
            print(chr(ord(content[i])-100), end="")
            
    else:
        break
```

### CSV 형식
- csv(comma, separated value) : 표 형태의 데이터를 쉼표로 구분

```python
import csv

with open("./sample1.csv","r") as f:
    reader = csv.reader(f)
    print(reader)

    next(reader) # 헤더명 제거 => ['번호', '이름', '가입일시', '나이']
    for c in reader:
        print(c)
```

```python
import csv

with open("./sample2.csv","r") as f:
    reader = csv.reader(f, delimiter="|") # |자리에 쉽표로 구분
    # print(reader)

    next(reader) # 헤더명 제거
    for c in reader:
        print(c)
```

```python
import csv

with open("./sample2.csv","r") as f:
    reader = csv.reader(f, delimiter="|")
    # print(reader)

    next(reader) # 헤더명 제거
    for row in reader:
        for col in row:
            print(col)
```

```python
import csv
# DictReader로 데이터 뽑기

with open("./sample1.csv","r") as f:
    reader = csv.DictReader(f)
    print(reader)

    next(reader) # 헤더명 제거
    for c in reader:
        for k,v in c.items(): # {"key" : "value"}
            print(k,v) # key value
```
